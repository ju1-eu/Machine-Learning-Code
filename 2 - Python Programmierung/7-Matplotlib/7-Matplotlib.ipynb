```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = np.array([8, 10, 22, 51])
```

```python
plt.plot(x, y, color="red")
plt.xlabel("X-Werte")
plt.ylabel("Y-Werte")
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```
